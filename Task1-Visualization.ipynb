```python
#Data visualization
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Plot a simple line graph with x = [1, 2, 3, 4] and y = [10, 20, 25, 30]
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()

```

```python
# 2. Add a title, x-axis label, and y-axis label to a plot in Matplotlib
plt.plot(x, y)
plt.title('Simple Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

```

```python
# 3. Create a bar plot using Seaborn for the dataset df with x='category' and y='values'
# Sample data
df = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D'],
    'values': [5, 10, 15, 20]
})
sns.barplot(x='category', y='values', data=df)
plt.show()

```

```python
# 4. Load the tips dataset in Seaborn and display the first 5 rows
tips = sns.load_dataset('tips')
print(tips.head())

```

```python
# 5. What function is used to display the plot in Matplotlib or Seaborn?
# The function used is `plt.show()` in Matplotlib, or `sns.show()` in Seaborn. This function displays the plot.
# Example:
plt.plot(x, y)
plt.show()

```

```python
# 6. Create a figure with 2 rows and 1 column of subplots, showing two different line graphs
fig, axes = plt.subplots(2, 1)  # 2 rows, 1 column
axes[0].plot(x, y)
axes[0].set_title('Line Graph 1')
axes[1].plot([1, 2, 3, 4], [5, 10, 15, 20])
axes[1].set_title('Line Graph 2')
plt.tight_layout()  # Adjust layout
plt.show()

```

```python
# 7. Add a legend to a plot in Matplotlib and specify its location (e.g., top right)
plt.plot(x, y, label='Line 1')
plt.legend(loc='upper right')  # Location of the legend
plt.show()

```
